```python
# https://stackoverflow.com/questions/47283197/geocoding-in-python-get-latitude-and-longitude-from-addresses-using-api-key

city_coordinates = {}
def get_coordinates(location):
    # Over limit error... hago que corra una sola vez a ver si lo puedo solucionar
    if not location in city_coordinates:
        location = g.geocode(location, timeout=15)
        city_coordinates[location] = "%s;%s" % (location.latitude, location.longitude)
    return city_coordinates[location]
    #if not location in city_coordinates:
    #    city_coordinates[location] = Geocoder.geocode(location)
    #return city_coordinates[location]


conversions_with_location['coordinates'] = conversions_with_location.apply(lambda row: get_coordinates(row['city']+", "+row['region']+", "+row['country']),axis=1)
conversions_with_location.sample(5)
```
